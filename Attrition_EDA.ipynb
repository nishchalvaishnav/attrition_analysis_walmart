```python
# What is the ATTRITION RATE of company?
attrition_count = df["Attrition"].value_counts()
```

```python
count_yes = attrition_count["Yes"]
count_no = attrition_count["No"]
total_employee = count_yes + count_no

attrition_rate = (count_yes/total_employee)*100
```

```python
print(f"Attrition rate = {attrition_rate:.2f}%")
```

```python
plt.pie(attrition_count)
```

```python
age_group = df.groupby('Age')['Attrition'].value_counts(normalize=True).unstack().fillna(0)
```

```python
print(attrition_rates)
```

```python

df["EnvironmentSatisfaction"].value_counts()
```

```python
satisfaction_groups = df.groupby('EnvironmentSatisfaction')['Attrition'].value_counts(normalize=True).unstack().fillna(0)
attrition_rates = satisfaction_groups['Yes'] * 100

print(attrition_rates)
```

# so here we can see greater the satisfaction lower the Attrition rate.

```python
attrition_rates_df = attrition_rates.reset_index()
attrition_rates_df.columns = ['Satisfaction', 'AttritionRate']

print(attrition_rates_df)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(attrition_rates_df['Satisfaction'], attrition_rates_df['AttritionRate'], color='blue')
plt.title('Job Satisfaction vs. Attrition Rate')
plt.xlabel('Job Satisfaction')
plt.ylabel('Attrition Rate')
plt.grid(True)
plt.show()
```

```python
attrition_col = pd.get_dummies(df["Attrition"])
attrition_col.head()
```

```python
df = pd.concat([df,attrition_col],axis=1)
df.head()
```

```python
df = df.drop(["Attrition","No"],axis=1)
df.head()
```

```python
df = df.rename(columns = {"Yes":"Yes=1 and No=0"})
df.head()
```


## so we can see male ratio is greater then females who are 

```python
sns.barplot(x="Gender",y="Yes=1 and No=0", data=df)
```

```python
df.groupby("Gender").nunique()
```

```python
female_yes_count = df[(df["Gender"]=="Female") & (df["Yes=1 and No=0"]==1)].shape[0]
print("female_yes_count",female_yes_count)
female_no_count = df[(df["Gender"]=="Female") & (df["Yes=1 and No=0"]==0)].shape[0]
print("female_no_count",female_no_count)
male_yes_count = df[(df["Gender"]=="Male") & (df["Yes=1 and No=0"]==1)].shape[0]
print("male_yes_count",male_yes_count)
male_no_count = df[(df["Gender"]=="Male") & (df["Yes=1 and No=0"]==0)].shape[0]
print("male_no_count",male_no_count)
```

```python
total_female_count = df[df["Gender"]=="Female"].shape[0]
print("total_female_count_in_company",total_female_count)
total_male_count = df[df["Gender"]=="Male"].shape[0]
print("total_male_count_in_company", total_male_count)
```

```python
plt.figure(figsize=(8,6))
plt.bar(["Female","Male"],[total_female_count,total_male_count],color=["pink","blue"])
plt.title("Total count of Male and Female Employees")
plt.xlabel("Gender")
plt.ylabel("Total count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.bar(["Yes","No"],[female_yes_count,female_no_count],color=["blue","red"])
plt.title("Female Employees Attrition 'Yes' vs 'No'")
plt.xlabel("Attrition")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.bar(["Yes","No"],[male_yes_count,male_no_count],color=["blue","red"])
plt.title("male Employees Attrition 'Yes' vs 'No'")
plt.xlabel("Attrition")
plt.ylabel("Count")
plt.show()
```

```python
female_yes_ratio = female_yes_count/total_female_count
print(f"female yes ratio: {female_yes_ratio:.2f}")

female_no_ratio = female_no_count/total_female_count
print(f"female NO ratio: {female_no_ratio:.2f}")
```
